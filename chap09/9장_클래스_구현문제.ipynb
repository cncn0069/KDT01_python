class Preprocessor:
    def normalize(self, data):
        min_val = min(data)
        max_val = max(data)
        return [(x - min_val) / (max_val - min_val) for x in data]

    def standardize(self, data):
        import statistics
        mean = statistics.mean(data)
        std = statistics.stdev(data)
        return [(x - mean) / std for x in data]
