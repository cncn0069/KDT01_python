```python
import numpy as np

def f_pure_python(self,val):
    return 2*val + 1
```

```python
def f_numpy(val):
    return np.array(2*val + 1)
```

```python
x = range(-5,6,1)
y = [f_pure_python(val) for val in x]
```

```python
x = np.arange(-5,6,1)
y = np.array(f_numpy(x))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(x,y)
#ax.axis([1,10,3,15])
plt.show()
```

```python
def f_pure_python(val):
    return 2*(val**2) - 3*(val) + 4
```

```python
def f_numpy(val):
    return np.array(2*(val**2) - 3*(val) + 4)
```

```python
x = range(-5,6,1)
y = [f_pure_python(var) for var in x]
```

```python
class String:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    
    def add(self)->int:
        return self.a+self.b

a = 1
b = 2
s = String(a,b)

r = s.add()
print(r)
```

```python
class DataLoader:
    def __init__(self, data):
        self.data = data
    
    def get(self, i:int):
        return self.data[i]
    
class MyData(DataLoader):
    def __init__(self, data):
        super().__init__(data)

    @staticmethod
    def get(self, i:int):
        return self.data[i]*10


data = range(10)
loader = MyData(data)

print(loader.get(1))

    
```
