파이썬함수 > zip(), map() 함수 공부하기 

```python
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

zipped = zip(list1, list2)# 짝을 tuple로 만드는 iterator 객체를 준다
print(zipped)
#zip을 하고 나서 list로 변경해야함
result = list(zipped)

print(result)  # [(1, 'a'), (2, 'b'), (3, 'c')]
```

```python
zipped = [(1, 'a'), (2, 'b'), (3, 'c')]
unzipped = zip(*zipped) ## 변수 unzipped는 iterator 객체이다 >> 2개의 iterable tuple로 구성됨

#print(tuple(unzipped))
### *zipped는 unpack하는 것으로(1,2,3), ('a','b','c')으로 나눈다::(1,'a')을 unpack하면 1, 'a'로 나눈다
### zip(*zipped)는 (1,2,3) 형태로 모은다 
## iterable 객체는 consume을 하기 때문에

t1, t2 = map(list, unzipped)
print(t1)
print(t2)
#여기서는 unzipped 인자들이 모두 consume 됨
list1, list2 = map(list, unzipped)
print(list1)  # [1, 2, 3]
print(list2)  # ['a', 'b', 'c']

```

```python
zipped = [(1, 'a'), (2, 'b'), (3, 'c')]

#a = *zipped 에러
#zip이 있어야함

print(zip(*zipped))

numbers, names = zip(*zipped)

for num, name in zip(numbers, names):
    print(f"num={num}, name= {name}")
```

```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4]
squared = map(square, numbers)

print(list(squared))  # [1, 4, 9, 16]
```

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)

print(list(squared))  # [1, 4, 9, 16]
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = map(lambda x, y: x + y, list1, list2)
print(list(result))  # [5, 7, 9]
```

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n* fact(n-1)
    
print(fact(5))
```

```python
def fibo(n):
    if n==0:
        return 1
    elif n ==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
    
print(fibo(5))
```
